```python
def remove_int(x):
    return ''.join(filter(str.isalpha, x))
```

```python
new_data['Districts']=new_data['Districts'].apply(remove_int)
```

```python
map_data=pd.read_json('map_data.json')
map_data
```

```python
import folium
```

```python
locations = map_data[['latitude', 'longitude']]
districts = locations.values.tolist()
districts
```

```python
map = folium.Map(location=[27.7172, 85.3045], tiles='Stamen Toner', zoom_start=100)
for point in range(0, len(districts)):
    folium.CircleMarker(districts[point], radius=8, color='red', fill=True, fill_color='red', opacity = 0.4).add_to(map)
map
```
